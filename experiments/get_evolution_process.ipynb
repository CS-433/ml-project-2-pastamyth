```python
import pandas as pd
from core.insilico_exps import ExperimentEvolution
from transformers.tf_exp import TransformerEvolution
```

```python
# Evolution for CNN
model_unit_0 = ("alexnet", ".classifier.Linear6", 0)
exp_evolution_0 = ExperimentEvolution(model_unit_0, explabel='alexnet_exp', savedir='./results/scores/evolution')
```

```python
# Run evolution
exp_evolution_0.run()
```

```python
fig_cnn = exp_evolution_0.visualize_trajectory(save=True, type='CNN')
```

```python
cnn_avg_scores, cnn_max_scores = exp_evolution_0.get_all_scores()
```

```python
# Save the scores into a csv file
import pandas as pd
df_cnn = pd.DataFrame(cnn_avg_scores, columns=['avg_scores'])
df_cnn['max_scores'] = cnn_max_scores
df_cnn.to_csv('./results/scores/evolution/alexnet_exp.csv')
```
